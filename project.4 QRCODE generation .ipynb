```python
import qrcode
from PIL import Image

# 1. Create a QR code object with custom settings
qr = qrcode.QRCode(
    version=1, # Version 1 is the smallest, version increases with data capacity (1 to 40)
    error_correction=qrcode.constants.ERROR_CORRECT_H, # High error correction (up to 30% errors corrected)
    box_size=10, # Size of each 'box' (pixel) in the QR code
    border=4, # Thickness of the border (minimum is 4)
)

# 2. Add data to the QR code
data = "priyanka harle"
qr.add_data(data)
qr.make(fit=True) # Adjusts size to fit data efficiently

# 3. Generate the image with custom colors
# make_image() returns a PIL Image object
img = qr.make_image(fill_color="darkblue", back_color="lightgray")

# 4. Save the image
img.save("custom_qrcode.png")
print("Custom QR code generated and saved as custom_qrcode.png")
```
