# HANGMAN GAME PROJECT

```python
import random

words = ["apple", "mango", "banana", "orange", "grapes", "cherry", "peach"]


secret_word = random.choice(words)
word_length = len(secret_word)


chances = word_length + 2

guessed_letters = []

display_word = ["_"] * word_length

print(" Welcome to Hangman!")
print("Guess the fruit name")
print("You have", chances, "chances")

while chances > 0:
    print("Word:", " ".join(display_word))
    guess = input("Enter a letter: ").lower()


    if guess in guessed_letters:
        print(" You have already guessed this letter. Try another.")
        continue  

    
    guessed_letters.append(guess)

    if guess in secret_word:
        print(" Correct guess ")
        for i in range(word_length):
            if secret_word[i] == guess:
                display_word[i] = guess
    else:
        chances -= 1
        print(" Wrong guess!")
        print("Remaining chances:", chances)

    
    if "_" not in display_word:
        print(" Congratulations! You guessed the word:", secret_word)
        break


if chances == 0:
    print(" Game Over! The word was:", secret_word)
```
